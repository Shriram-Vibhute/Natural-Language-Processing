```python
# Meta Characters
test_string_1 = "123abc.456789abc123defGHI.ABC \n xyz345WXYZ"
test_string_2 = "The cat sat on the mat with a bat and a hat."
test_string_3 = "The cat sat on the mat with a bat and a hat. $at, 8at, at."
text = """Hello World!
This is a test.
Hello again!
Python is fun.
Hello Python"""
```

```python
# . meta character is used to match any character except \n (new line)
pattern = re.compile(r".")
re.findall(pattern=pattern, string=test_string_1)
```

```python
# ^ meta character used to match the pattern with the starting of the string else None
pattern = re.compile(r"^Hello")
re.findall(pattern, text)
```

```python
# $ meta character used to match the pattern with the ending of the string else None
pattern = re.compile(r"Python$")
re.findall(pattern, text)
```

```python
# Character Sets - refers to a combination of characters
test_string = "123abc.456789abc123defGHI.ABC \n xyz345WXYZ"

# Custom Character Sets
# [] - This character set search for every single individual character in the string
pattern = re.compile(r"[aeiouAEIOU]") # Finding all the vowels in the string
iterator = re.finditer(pattern, test_string)
for match in iterator:
    print(match)
```

```python
# Character Ranges - refers to a range of characters
pattern = re.compile(r"[a-zA-Z0-9]") 
iterator = re.finditer(pattern, test_string)
for match in iterator:
    print(match)
```

```python
# Character Negation - refers to a set of characters that are not in the set
pattern = re.compile(r"[^aeiouAEIOU]") # Finding all the non-vowel characters in the string
iterator = re.finditer(pattern, test_string)
for match in iterator:
    print(match)

# NOTE: The meaning of ^ outside the character set means "not" | The meaning of ^ inside the character set means "starting with"
```

```python
pattern = re.compile(r"^[aeiouAEIOU]") # [] looses its meaning
iterator = re.finditer(pattern, test_string)
for match in iterator:
    print(match)
```

```python
# Predefined Character Sets - Do not use square brackets to use them otherwise they considered as custom character sets
# \d - Any digit character
# \D - Any non-digit character
# \s - Any whitespace character
# \S - Any non-whitespace character
# \w - Any word character
# \W - Any non-word character
# \b - Any word boundary character
# \B - Any non-word boundary character

pattern = re.compile(r"[\d\s]") # Finding all the digit and whitespace characters in the string
iterator = re.finditer(pattern, test_string)
for match in iterator:
    print(match)
```

```python
pattern = re.compile(r"\bxyz") # Finding all the word boundary characters in the string
iterator = re.finditer(pattern, test_string)
for match in iterator:
    print(match)
```

```python
# Quantifiers - To specify the number of times a character or a group of characters should occur
# * - Zero or more times
# + - One or more times
# ? - Zero or one time
# {n} - Exactly n times
# {n,} - At least n times
# {n,m} - Between n and m times

test_string_1 = "no no noo nooo noooothing noo"

test_string_2 = "The colour of the sky and the color of the ocean are different."

test_string_3 = "We can use either http or https to generate URLs."

test_string_4 = "aaXaaaXaaaaXaaaaaX"

test_string_5 = "<div>First div</div><div>Second div</div>"
```

```python
pattern = re.compile(r"noo*") # o* - o zero or more times
iterator = re.finditer(pattern, test_string_1)
for match in iterator:
    print(match)
```

```python
pattern = re.compile(r"noo+") # o+ - o one or more times
iterator = re.finditer(pattern, test_string_1)
for match in iterator:
    print(match) # There is no `no`
```

```python
pattern = re.compile(r"colou?r") # u? - u may or may not occur
iterator = re.finditer(pattern, test_string_2)
for match in iterator:
    print(match)
```

```python
pattern = re.compile(r"https?") # s? - s may or may not occur
iterator = re.finditer(pattern, test_string_3)
for match in iterator:
    print(match)
```

```python
pattern = re.compile(r"aa{2}") # s? - s may or may not occur
iterator = re.finditer(pattern, test_string_4)
for match in iterator:
    print(match)
```

```python
pattern = re.compile(r"a{2,5}") # From higher occurance to lower occurance (5, 4, 3, 2) - both inclusive
matches = re.finditer(pattern, test_string_4)
for match in matches:
	print(match)
```

*If you dont mention the `?` after the greedy quantifier, it will try to match as many times as possible. If you mention the `?` after the greedy quantifier, it will try to match as few times as possible.*

```python
pattern = re.compile(r"noo*?") # As few as possible
iterator = re.finditer(pattern, test_string_1)
for match in iterator:
	print(match)
```

```python
# Combination of all
pattern = re.compile(r"<\w+>.*?</\w+>") # As few as possible
matches = re.findall(pattern, test_string_5)
matches
```

```python
pattern = re.compile(r"<\w+>.*</\w+>") # As many as possible
matches = re.findall(pattern, test_string_5)
matches
```
